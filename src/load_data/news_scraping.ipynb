```python
import requests

url = "https://finviz.com/news.ashx?v=3"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} # To bypass the Just a moment, cookies window
response = requests.get(url, headers=headers)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())  # prints well-formatted HTML
```

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# Now for each news, I can gather the link (a class="nn-tab-link" href=link), 
# the title (<a class (...)>title</a> , 
# the ticker (a class="fv-label stock-news-label is-opaque is-!negative!-50" href="(...)=ticker" . !can change!
```

```python
from datetime import datetime

links = []
titles = []
tickers = []
times = []

now = datetime.now

for e in soup.find_all('tr', class_='news_table-row'):
    subsoup_links = e.find('a', class_='nn-tab-link')
    link = subsoup_links.get('href')
    # Sometimes the links begin with '/' bc the news website in finviz
    if (link[0] == '/'):
        link = f"https://finviz.com{link}"
    links.append(link)
    
    subsoup_titles = e.find('a', class_='nn-tab-link')
    #print(subsoup_titles.get_text())
    title = subsoup_titles.get_text()
    titles.append(title)

    # find_all so I can get all the tickers
    subsoup_tickers = e.find_all('a', class_='stock-news-label')
    #print(f"AAAAAA : {subsoup_tickers}\n\n")
    tickers_list = []
    for ticker_class in subsoup_tickers:
        #print(f"TICKER in subsoup : {ticker} \n\n")
        ticker_text = ticker_class.get('href')
        #print(ticker_text)
        ticker = ticker_text.split('=')[1]
        tickers_list.append(ticker)
    tickers.append(tickers_list)
    print(f"link : {link} \n title : {title} \n tickers : {tickers_list} \n")
    # print(e)
    # print("\n")
    # print("=====================\n\n")

print(f"{len(links)} links found. here you go : \n{links}\n\n")
print(f"{len(titles)} titles found. here you go : \n{titles}\n\n")
print(f"{len(tickers)} tickers found. here you go : \n{tickers}\n\n")
```

```python
import pandas as pd
```

```python
data = {
    'link': links,
    'title': titles,
    'tickers': tickers,
    'date_time':times
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('data.csv', mode='a', index=False, header=False)
```
